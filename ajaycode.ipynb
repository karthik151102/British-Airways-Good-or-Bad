```r
# Load necessary libraries
library(ggplot2)
library(dplyr)
library(readr)

# Load the dataset (Assuming advertising.csv is in the working directory)
advertising <- read.csv("advertising.csv")

# View the first few rows
head(advertising)

# Check structure and summary statistics
str(advertising)
summary(advertising)

# Check for missing values
sum(is.na(advertising))

# Compute correlation matrix
cor_matrix <- cor(advertising[, sapply(advertising, is.numeric)], use = "complete.obs")
print(cor_matrix)

# Perform Linear Regression (Predicting Sales based on TV, Radio, and Newspaper)
model <- lm(Sales ~ TV + Radio + Newspaper, data = advertising)
summary(model)

# Visualizing relationships using scatter plots
ggplot(advertising, aes(x=TV, y=Sales)) +
  geom_point() +
  geom_smooth(method="lm", col="red") +
  ggtitle("Sales vs TV Advertising")

ggplot(advertising, aes(x=Radio, y=Sales)) +
  geom_point() +
  geom_smooth(method="lm", col="blue") +
  ggtitle("Sales vs Radio Advertising")

ggplot(advertising, aes(x=Newspaper, y=Sales)) +
  geom_point() +
  geom_smooth(method="lm", col="green") +
  ggtitle("Sales vs Newspaper Advertising")

# Checking residuals of the regression model
par(mfrow = c(2, 2))
plot(model)

# ANOVA Test
anova(model)
```
